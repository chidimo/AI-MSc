```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('data/Social_Network_Ads.csv')
df.describe(include='all')
```

```python
sns.histplot(df.Age)
plt.title('Age distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Product purchasing based on the age')
sns.histplot(x='Age', hue='Purchased', data=df)
plt.show()
```

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df
```

```python
X = df.iloc[:, [1,2,3]].values
X
```

```python
y = df.iloc[:, 4]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## K Nearest Neighbor

```python
SC = StandardScaler()
X_train_s = SC.fit_transform(X_train)
X_test_s = SC.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train_s, y_train)
```

```python
y_pred = classifier.predict(X_test_s)
y_pred
```

```python
print(y_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: %.2f\n\n' %acc)
cm = metrics.confusion_matrix(y_test, y_pred)
print('Confusion Matrix')
print(cm, '\n\n')
print('-----' * 25)
result = metrics.classification_report(y_test, y_pred, zero_division=1)
print('Classification Report: \n')
print(result)

ax = sns.heatmap(cm, cmap='flare', annot=True, fmt='d')

plt.xlabel('Predicted Class', fontdict={'size': 12})
plt.ylabel('True Class', fontdict={'size': 12})
plt.title('Confusion Matrix', fontdict={'size': 12})
plt.show()
```

## Decision Tree

```python
tree_classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree_classifier.fit(X_train, y_train)

y_pred = tree_classifier.predict(X_test)
y_pred

acc = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: %.2f\n\n' %acc)
cm = metrics.confusion_matrix(y_test, y_pred)
print('Confusion Matrix')
print(cm, '\n\n')
print('-----' * 25)
result = metrics.classification_report(y_test, y_pred, zero_division=1)
print('Classification Report: \n')
print(result)

# draw heatmap
ax = sns.heatmap(cm, cmap='flare', annot=True, fmt='d')

plt.xlabel('Predicted Class', fontdict={'size': 12})
plt.ylabel('True Class', fontdict={'size': 12})
plt.title('Confusion Matrix', fontdict={'size': 12})
plt.show()
```
